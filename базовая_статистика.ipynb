открываем набор данных о состоянии лошадей, все ? заменяем на NaN

```python
import pandas as pd
df=pd.read_csv('horse_data.csv', na_values='?')
```

```python
df.head(40)
```

Выбираем несколько столбцов из набора

```python
df2=df[['2', '1', '38.50', '66','28','3','11300']]
```

Даем названия столбцам

```python
df2.columns=['surgery', 'age','rectal temperature','pulse','respirat.rate','temp.extremit.','type']
```

```python
df2.head(1000)
```

Меняем тип данных столбца

```python
df2['rectal temperature'].astype(float)
```

считаем среднюю температуру без выбросов 

```python
df2[(df2['rectal temperature']>=35.00)&(df2['rectal temperature']<=40.00)]['rectal temperature'].mean()
```

формирую датафрейм без выбросов по столбцу rectal temperature через межквартильный размах

```python
q1=df2['rectal temperature'].quantile(0.25)
q3=df2['rectal temperature'].quantile(0.75)
igr=q3-q1
low= q1-(1.5*igr)
up=q3+(1.5*igr)
rem= df2[df2['rectal temperature'].between(low,up,inclusive=True)].sort_values('rectal temperature')
rem
```

нахожу эти выбросы ( keep =False только уникальные значения, дублированные удаляются)

```python
pd.concat([df2,rem]).drop_duplicates(keep=False)
```

количество пропусков

```python
df2.isna().sum()
```

Заменяю пропуски медианой в разрезе temp.extremit. (так как есть корреляция)

```python
df2['rectal temperature'].fillna(df2.groupby('temp.extremit.')['rectal temperature'].transform('median'),inplace=True)
```

```python
df2['rectal temperature'].fillna(df2['rectal temperature'].mean(),inplace=True)
```

Заменяю пропуски медианой в разрезе rectal temperature (так как есть корреляция)

```python
df2['temp.extremit.'].fillna(df2.groupby('rectal temperature')['temp.extremit.'].transform('mean'),inplace=True)
```

```python
df2.isna().sum()
```

Заменяю пропуски медианой в разрезе температуры конечностей (так как есть корреляция)

```python
df2['pulse'].fillna(df2.groupby('temp.extremit.')['pulse'].transform('median'),inplace=True)
```

```python
df2.isna().sum()
```

Заменяю пропуски средним в разрезе пульса (так как есть корреляция)

```python
df2['respirat.rate'].fillna(df2.groupby('pulse')['respirat.rate'].transform('mean'),inplace=True)
```

```python
df2.isna().sum()
```

```python
df2['surgery'].fillna(df2.groupby('respirat.rate')['surgery'].transform('median'),inplace=True)
```

```python
df2.isna().sum()
```

после заполнения пропусков с помощью группировки все равно осталось несколько . Поэтому заменяем их без группировки 

```python
df2['respirat.rate'].fillna(df2['respirat.rate'].mean(),inplace=True)
```

```python
df2['temp.extremit.'].fillna(df2['temp.extremit.'].mode()[0],inplace=True)
```

```python
df2.isna().sum()
```
